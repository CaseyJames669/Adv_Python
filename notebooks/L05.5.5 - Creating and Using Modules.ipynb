## Module contents  
### Module description:  
> The module description should help a reader determine if the module contains the tools necessary to be helpful in a particular programming task.  It should be located at the top of the module, before function / class definitions, and describe the contents in aggregate.  The descriptions of the individual functions and/or classes should be left for their definitions.  

> In our example, we may have something like:

### Functions and function descriptions:  
> As we have already seen, we can use **docstrings** to provide **user-level documentation** in the function definitions, themselves.  This documentation should follow the specification laid out in **PEP 257**.  
I require that each function have:
>> * **"Short description"**, consisting of no more than one line of less than 80 spaces, in the form of a command  
* a blank line  
* an optional **Purpose** description if the short description needs clarification / augmentation,
* **Preconditions** describing what the function needs from the "user" (i.e. application programmer) in order to do its job.  Minimally, this means that requirements of/for the parameters in the parameter list must be described.  
* **Postconditions** describing what the "user" can expect of the function.  This should address the returned value(s) and any side effects that the function produces.  
* optional **doctests** which serve two purposes:  to demonstrate how to use the function and to act as unit tests for the function.  

> Of course, the function definitions also provide the code necessary to perform the task.  This code should be adequately documented in a block commenting style.  Inline comments should be kept to a minimum or avoided altogether.  

### Unit test code  
> Each function should be thoroughly tested.  Rather than writing test code, testing the function, and then deleting the test code, the code is "sequestered" with an if statement so that it is not executed when the module is imported, but only when the module is run directly (as a python script).  By maintaining the code in the module file, the functions may be retested after subsequent changes are made to the module.

```python
import string
def word_strip(wd):
    '''Strip string and return lowercase version of result
  
    Return a lowercase version of the word with all non-alphabetic
    symbols stripped away from the boundaries, but with such symbols
    remaining in the interior of a word.
  
    Precondition: w is a string
    Postcondition: returns 'stripped' string in lowercase

    Typical examples of this behavior include:
      
    >>> word_strip("don't")
    "don't"
    >>> word_strip('end.')
    'end'
    >>> word_strip('jack-o-lantern')
    'jack-o-lantern'
    >>> word_strip('"Stop!"')
    'stop'
    >>> word_strip('&%^$*%*%"')
    ''
    >>> word_strip('hello there!!')
    'hello there'
    '''
    
    # locate first desirable character
    for start, ch in enumerate(wd):
        if ch.isalpha():
            break

    # locate last desirable character
    end = len(wd)-1
    while end >= 0 and wd[end] not in string.ascii_letters:
        end -= 1

    # return an intentionally lowercase stripped word
    return wd[start:end+1].lower()
```

```python
import re

def words(data = None):
    '''Return an iterator over data that returns tuples of (line, word)
    
    Preconditions:  'data' must be an iterable of strings
    Postcondition:  returns an iterator over data that returns the a tuple
                    (i, w) where:
                        w is the stripped lower-case representation of next
                            word in data and
                        i is the index of the item in data from which the 
                            word w was retrieved.
                    A stripped word is one that begins and ends with an 
                    alphabetic character and has no intervening white space.
    '''
    # Your code here
    data = " ".join(re.findall("[a-zA-Z]+", data)).lower()
        
    iter_matches = enumerate(re.finditer(r'\w+', data))
    for i,match in iter_matches:
        yield(i,match.group(0))
```

```python
import re

def word_freq(data):
    '''Return a dict of {word:frequency} of words in data
    
    Precondition:  data must be an iterable of strings
    Postcondition:  returns a dict of {word:frequency} where
                    word is a lower case representation of a word 
                    from data and frequency is the word's frequency
                    of occurance in data
                    
    '''
    words = [word.lower() for line in data for word in re.split(r'\W+', line)]
    words = dict((i,words.count(i)) for i in words)
    print(words)

    pass
```

## Using a module  
Once the module has been constructed and tested, it can be placed into a library of modules to be used by application programs.  Those programs may make use of the modules by using the **import** command.  

### Where to place modules  
> When Python is installed on a system, it is configured to look in a specific folder for library modules.  Often, however, regular users cannot add modules to this folder, so other measures must be taken.  
Python will search for modules in various folders (in order) according to the following rules:  
>> * the current folder,
* the folders whose paths are contained in the user's (bash) shell PYTHONPATH environmental variable, and
* in the path(s) configured at installation time.
