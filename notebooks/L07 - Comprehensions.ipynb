## Set builder notation

### Description

In mathematics we use a special notation, called **set builder notation**, to indicate the contents of a set.  The syntax is, roughly, 

> S = {f(x) | x has some particular property}

Where

* S is the name of the set being constructed,
* {, } are symbols used to indicate that the result is a set,
* the property is described using accepted mathematical notation,
* x stands for a general value satisfying the property,
* | is a symbol you can interpret as "such that", and
* f(x) is an expression, involving the value x, which describes the element to be added to the set.


### Some Set Builder Notation Examples

* **N** = the set of natural numbers = {x | x in 0, 1, 2, 3, ...}
* **I** = the set of all (positive and negative) integers = {+/- x | x in N}
* **P** = the set of positive integers = {x | x in I, if x > 0}
* **E** = the set of even integers = {2*x | x in I}
* **E** = the set of even integers = {x | x in I, if x%2 == 0}

## Python Comprehensions

Python has adopted an approach similar to the mathematics set builder notation, called **comprehension**, to provide a concise way to create new lists, sets, and dicts.  Commonly, the elements of the new structures are the result of some operations applied to each member of an existing sequence or iterable, or to a subsequence of those elements that satisfy a certain condition.
Once you have mastered the basics of list comprehensions, you can apply those same techniques to creating set and dict comprehensions.  You should find that comprehensions are more legible and natural than their multiline looping counterparts.  Furthermore, in many instances, the comprehensions are more efficient with regard to execution time and memory requirements.

### Lists

A **list comprehension** consists of **brackets** containing an **expression** followed by a **for clause**, then **zero or more for or if clauses**.  The result will be a new list resulting from evaluating the expression using values generated by the clauses.

We'll use some examples to make the description clear.
Let's consider the list of names:

```python
names = ['wilma', 'fred', 'barney', 'betty']
print(names)
```

**Example 1.**

Suppose we want to create a new list consisting of capitalized versions of the strings in *names*.  

**Solution 1.1**  
We could write a loop to accomplish this:

```python
names = ['wilma', 'fred', 'barney', 'betty']
cap_names = []
for name in names:
    cap_names.append(name.capitalize())
print(cap_names)
```

**Solution 1.2**  
Now, with our understanding of set builder notation, we can write a *list comprehension* ... using Python syntax, of course.  In our set builder notation, the above code could look like

> cap_names = {capitalize(x) | x in names}

We can do the same in Python:

```python
names = ['wilma', 'fred', 'barney', 'betty']
cap_names = [x.capitalize() for x in names]
print(cap_names)
```

**Example 4.**

In the above examples, the source for the elements of our new list was itself a list.  This is not necessary.  The source must be an **iterable** ... that is anything from which we can get a *next* element.

Suppose we want to construct a list containing the squares of the first 5 positive integers.  We could write the code

    squares = []
    for num in range(1, 6):
        squares.append(num ** 2)

We can also perform the same task in a list comprehension:

```python
squares = [x ** 2 for x in range(1, 6)]
print(squares)
```

**Example 5.**

Having multiple *for* clauses in the 'properties' portion of the comprehension has an effect identical to nesting *for* loops.  This allows us to draw values from more than one iterator when constructing the elements of our result.  
For example, suppose that we want to form all twelve ratios with numerator coming from the list [1, 2, 3] and denominator from the list [11, 12, 13, 14].  
We could employ the following python nested loop:

    ratios = []
    for x in range(1, 4):
        for y in range(11, 15):
            ratios.append(str(x) + '/' + str(y))

Or, we could construct an elegant list comprehension.  Note that the simple linear, left-to-right flow of the code makes the list comprehension easy to read and understand. 

```python
ratios = [str(y) + '/' + str(x) for x in range(1, 4) for y in range(11, 15)]
print(ratios)
```

Note that multiple 'for' clauses are treated as if they were nested ... with the leftmost 'for' producing the outer-most loop.

In the previous Example and Exercises 4 and 5, we generated all the pairs of integers that could be formed with the first coming from one iterable and the second coming from another iterable.  In many situations, we may want to pair up only the elements in corresponding positions of a pair of sequences. Consider a situation where we wish to find percentage grades.

**Example 7.**

Suppose that *scores* is a list of a student's assignment scores:
> scores = [18, 29, 14, 31]

and *possible* is a list of maximum possible scores on the assignments:
> possible = [24, 30, 25, 35].

The student scores and maximum scores correspond positionally ... that is scores[0] is out of possible[0] points, etc.
Our goal is to create a list of percentage scores for the assignments.  Of course, a standing goal is to do it elegantly.

Assuming that the two lists have the same length, one approach is:

    >>> scores = [18, 29, 14, 31]
    >>> possible = [24, 30, 25, 35]
    >>> percents = []
    >>> for i in range(len(scores)):
    ...     ratio = scores[i] / possible[i]
    ...     percents.append(ratio)
    ... 
    >>> print(percents)
    [0.75, 0.9666666666666667, 0.56, 0.8857142857142857]

There are at least two unnecessary aspects that make this solution less than elegant, however.

* First, we have introduced the loop variable *i*, which is not really part of the problem and is just one more thing for readers to absorb.
* Next, the code spans several lines and includes two "levels of logic".  Neither of these helps to make the code 'natural' to read.

Enter list comprehension ... with a little help from *zip*.  If you haven't heard of or used *zip* just type 'zip' into the Quick Search box of the Python documentation for a description and examples of this nice little tool that creates tuples of elements from two or more iterables.  You can also type 'help(zip)' in a Python shell to get its pydoc synopsis.

In the comprehension below, *zip* forms a iterable of tuples with first element from *scores* and second element from the corresponding position of *possible*.  Note that the new iterable is NOT a list and must be iterated over to see its values.

Try executing:

```python
scores = [18, 29, 14, 31]
possible = [24, 30, 25, 35]
pairs = zip(scores, possible)
print(pairs)
```

Not very helpful, is it.  Now, try a list comprehension to make a list of the tuples:

```python
scores = [18, 29, 14, 31]
possible = [24, 30, 25, 35]
pairs = [pair for pair in zip(scores, possible)]
print(pairs)
```

OK, that got us the pairs we wanted ... now we just need to evaluate the ratios.  We can 'unpack' the tuples right in the 'for' clause and then evaluate the ratio of the two elements.

```python
scores = [18, 29, 14, 31]
possible = [24, 30, 25, 35]
percents = [x/y for x, y in zip(scores, possible)]
print(percents)
```

### Sets

Recall, a **set** is an unordered collection of unique elements.  We can use the comprehension techinques we employed for lists to construct sets, as well.  The difference in syntax is we use { and } to denote that we are constructing a set, instead of [ and ].

**Example 8.**  

Let's find all the fractions in lowest terms that can be formed with a integer numerator between 1 and 100 inclusive and with integer denominator between 1 and 100 inclusive.  
Remember that a fraction is in lowest terms if it's numerator and denominator have no factors in common (other than 1).  
We can **reduce a fraction to lowest terms** by dividing numerator and denominator by their **greatest common divisor**.  

Obviously if we find all these fractions, some will be duplicated:  '1/4' and '2/8' will both yield the value '1/4' ... you can find others.  
So, we can build a **set** to contain the values.

```python
from fractions import gcd
lowest_terms = {str(x // gcd(x,y)) + '/' + str(y // gcd(x,y)) for x in range(1, 5) for y in range(1, 10)}
print(lowest_terms)
```

This code has some features worth noting:  

> * The **fraction.gcd()** function - a rule of thumb should be "if it's useful, there is probably a module for that".  Programmers ococcasionally must work with fractions (and **greatest common divisor** is useful in that context).  A quick search for **gcd** in the Python documentation leads us to the **fractions** module.
* Multiple **for** clauses - when multiple for clauses are present, they are treated as being nested, with the left-most **for** creating the outermost loop and the right-most **for** being the innermost loop.
* The fractions aren't listed in an order related to their size or in the order in which they were generated.  **set** uses a mapping strategy similar to what **dict** uses.

Here is the looping structure that the above comprehension supplants:

```python
lowest_terms = set()
for x in range(1, 5):
    for y in range(1, 10):
        lowest_terms.add(str(x // gcd(x,y)) + '/' + str(y // gcd(x,y)))
print(lowest_terms)
```

### Dictionaries (dict)

This discussion assumes that you at least know what dicts are about.  If not, take a look at the section on Dictionaries in the Tutorial section of the Python documentation or look 'dict' up in its Built-Ins section.

As with set comprehension, dict comprehension semantics mirror that of list comprehension.  The syntactic difference is:

* we use { and } to instead of [ and ] to indicate ... well ... that we are not creating a list (remember that we use { and } for sets, too)
* we supply 'key : value' pairs for the items in the dict (this syntax distinguishes our dict from the set)

**Example 9.**



We have parallel lists - one of building names and one of the heights of the corresponding buildings, in meters.  We wish to create a dict with key:value pairs being building:height.

```python
buildings = ['Burj Khalifa', 'Eiffle Tower', 'Empire State', 'Sears']
heights = [829.8, 275, 381, 412.4]
tall_buildings = {b:h for b, h in zip(buildings, heights)}
print(tall_buildings)
```
