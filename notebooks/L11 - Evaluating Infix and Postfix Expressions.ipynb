## Infix and Postfix notations  
Arithmetic expression notations are categorized by the position of the operator relative to the operands.
> * In **infix**, operators are **between** the operands, with operator presidence, associativity, and grouping with '(' and ')' determining the order of operations.  
> * In **postfix** notation, operators **follow** the operands (hence post-) and the sequential ordering of the operands and operators determines the order in which operations are performed.  When an operator is encountered in a left-to-right scan of the expression, it is applied to the two operands immediately preceeding it.  The three tokens are then replaced by the result and the scan continues.  
> * In **prefix** notation, operators **preceed** the operands.  

In this lesson we are concerned with conversion of **infix** notation to **postfix** and evaluation of **postfix** expressions.  As we will see, **postfix** representation is convenient for algorithmic evaluation because of the low processing and storage requirements.

## Postfix Notation  
The history and use of **Postfix** (or **Reverse Polish - RPN**) notation is described in  
>> <a href="http://en.wikipedia.org/wiki/Reverse_Polish_notation" target="_blank"> Wikipedia - Reverse Polish Notation </a>  

There, you will find a discussion of the process and the benefits of using RPN in conjunction with a stack for efficient evaluation of arithmetic expressions involving binary operators.  

One great benefit of RPN is that expressions do not require parentheses to indicate order of operations - operator presidence, associativity, and order of operation (grouping) are fully specified by the order in which the operands and operators are listed in the expression.  The evaluation of the expression can be done with a queue and a stack, as follows:  

> while the expression queue is not empty  
>> remove an expression token from the front of the queue  

>> if the token is an operand (identifier or number),  
>>> push the token onto the stack  

>> else the token is an operator, so  
>>> call the operator token **OP**  
>>> pop the top of the stack into **op2**  
>>> pop the top of the stack into **op1**  
>>> perform the operation **op1 OP op2**  
>>> push the result onto the stack  

> the result is the only item left on the stack  

In our investigation of the process, we will focus on addition, subtraction, multiplication, division, and exponiation using the sybols '+', '-', '*', '/', and '^', resp.  
To simplify input processing, we will also require that the tokens in the expression be separated from each other by white space.

Consider the **infix** expression:  
> 50 / ( 3 ^ ( 1 + 2 ) - 2 ) * 12  

This **infix** expression can be written, using **postfix** notation, as  
> 50 3 1 2 + ^ 2 - / 12 *  

Using the above algorithm, evaluation procedes as:  
> in a left to right scan, numerical tokens are moved from the input queue to the operand stack until the first operator **'+'** is encountered:  
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;50 3 1 2  
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;^ 2 - / 12 *  
>> **OP:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;+  

> the top two elements on the stack are popped into **op2** and **op1**, resp, and **op1 OP op2** evaluates to **3**.  The result is pushed onto the stack to yield:  
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;50 3 **3**  
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;^ 2 - / 12 *  
>> **OP:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  

> In the next scan, **^** is removed from the input queue and triggers the evaluation process
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;50   
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;2 - / 12 *  
>> **op1:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;3  
>> **OP:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;^  
>> **op2:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**3**  

>  the operation **3 ^ 3** evaluates to **27**, and the configuration becomes:  
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;50 **27**  
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;2 - / 12 *  
>> **OP:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  

> Scanning resumes with **2** being removed from the queue and pushed onto the stack.  
Next, the operator **-** is removed from the queue.  
This triggers popping **2** into **op2** and **27** into **op1**.  
This yields the configuration:  
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;50   
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;/ 12 *  
>> **op1:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;27  
>> **OP:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-  
>> **op2:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2   

> **27 - 2** yields **25**, which is pushed onto the stack.  The next token removed from the queue is the operator **/**, which leads to:
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;50 **25**  
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;12 *  
>> **OP:**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;/  

> This triggers popping **25** into **op2** and **50** into **op1**.  
**50 / 25** is evaluated and  
**2** is pushed back onto the stack:  
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;**2**  
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;12 *  

> and finally, **12** is moved from queue to stack  
**\*** is removed from the queue, triggering  
**12** is popped from the stack to **op2** and **2** is popped from the stack into **op1**.  
**2 \* 12** evaluates to **24**, which is pushed onto the stack.
This leads to the configuration:
>> **Stack:**&nbsp;&nbsp;&nbsp;&nbsp;**24**  
>> **Queue:** &nbsp;&nbsp;&nbsp;&nbsp;**empty**  

> The result can be popped from the stack.  
Note: If the stack is not empty after this last pop, the expression was ill-formed.

## Infix to Postfix  
You may be thinking, "This is all real nice, but how do I even find the postfix expression for a given infix expression ... much less the 'want to' to actually do this?"  
The answer to both parts is, "There's an algorithm for that ... and where there's and algorithms there's a programming solution."  

The algorithm for translating **infix** expressions into **postfix**, developed by Edsger Dijkstra, is known as the **Shunting Yard Algorithm**.  It is particularly interesting to us at this point because it makes use of both the Stack and the Queue data structures.  
The algorithm is described and illustrated in the >> <a href="http://en.wikipedia.org/wiki/Shunting-yard_algorithm" target="_blank"> Wikipedia Shunting Yard Algorithm article</a>  

The algorithm is a little more complex than we need.  
The following algorithm filters out the steps / considerations we need:

#### Algoritm for infix to postfix expression representation.
    
    While there are tokens to be read:

        Read a token.
        If the token is a number, then add it to the output queue.

        If the token is an operator, o1, then:

            while (there is an operator token, o2, at the top of the stack, and

                    either o1 is left-associative and its precedence is equal
                            to that of o2, or
                    o1 has precedence less than that of o2),

                pop o2 off the stack, onto the output queue;

            push o1 onto the stack.

        If the token is a left parenthesis, then push it onto the stack.
        If the token is a right parenthesis:

            Until the token at the top of the stack is a left parenthesis, 
                    pop operators off the stack onto the output queue.
            Pop the left parenthesis from the stack, but not onto the output queue.
            If the stack runs out without finding a left parenthesis, 
                    then (ERROR) there are mismatched parentheses.

    When there are no more tokens to read:

        While there are still operator tokens in the stack:

            If the operator token on the top of the stack is a parenthesis,
                    then there are mismatched parentheses.
            Pop the operator onto the output queue.

    Output queue now contains the postfix representation of the input
            infix expression.
